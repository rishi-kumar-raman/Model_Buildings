## Data preparation for the model

```python
X=df1_data.drop(['target'],axis=1)
y=df1_data['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
def model_eval(algo,Xtrain,ytrain,Xtest,ytest):
    algo.fit(Xtrain,ytrain)
    y_train_prob=algo.predict_proba(Xtrain)[:,-1]
    y_train_pred=algo.predict(Xtrain)
    print('Accuracy score for train',accuracy_score(ytrain,y_train_pred))
    print(confusion_matrix(ytrain,y_train_pred))
    
    y_train_prob=algo.predict_proba(Xtest)[:,-1]
    y_train_pred=algo.predict(Xtest)
    print('Accuracy score for test',accuracy_score(ytest,y_train_pred))
    print(confusion_matrix(ytest,y_train_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification,accuracy_score,roc_auc_score,roc_curve,classification_report,confusion_matrix
```

```python
lr=LogisticRegression()
```

```python
model_eval(lr,X_train,y_train,X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
max_depth=[2,3,4,5,6,7,8]
min_samples_split=[2,3,4,5,6,7,8,9,10]
max_samples_leaf=[1,2,3,4,5,6,7,8,9]
criteria=['gini','entropy']

params={'max_depth':max_depth,'min_samples_split':min_samples_split,'min_samples_leaf':max_samples_leaf,'criterion':criteria}
gsearch=GridSearchCV(dtc,param_grid=params,scoring='roc_auc',cv=3)
gsearch.fit(X_train,y_train)
```

```python
gsearch.best_params_
```

```python
dtc=DecisionTreeClassifier(**gsearch.best_params_)
```

```python
dtc.fit(X_train,y_train)
```

```python
from IPython.display import Image  
from sklearn.externals.six import StringIO  
from sklearn.tree import export_graphviz
import pydotplus

features = X_train.columns
# Create DOT data
dot_data = export_graphviz(dtc, out_file=None, feature_names=features)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)  

# Show graph
Image(graph.create_png())
```

```python
model_eval(dtc,X_train,y_train,X_test,y_test)
```
